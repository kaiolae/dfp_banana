```python
import pandas as pd
training_dir = "../oct8_explicitly_controlled_battery_charging_agnostic/"
training_data = training_dir+"/dfp_stats.txt"
data = pd.read_csv(training_data, sep=" ")
```

```python
#Loss: not going anywhere! maybe due to the data changing as policy changes
data[['mavg_loss']].plot()
```

```python
#That looks promising. Initially, food and poison increases.
#Later, only food keeps increasing.
data['median_avg_score']= data['mavg_score'].rolling(11).median()
data['median_foods']= data['mavg_food'].rolling(11).median()
data['median_poisons']= data['mavg_poison'].rolling(11).median()
data['median_battery']= data['mavg_battery'].rolling(11).median()
data['median_battery_picks']= data['mavg_num_batteries'].rolling(11).median()
data[['median_avg_score', 'median_foods', 'median_poisons', 'mavg_num_batteries']].plot()
```
